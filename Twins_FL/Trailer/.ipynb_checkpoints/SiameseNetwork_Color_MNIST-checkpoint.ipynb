# Function

## Plots the given 'metric' from 'history'.

```python
def plt_metric(history, metric, title, has_valid=True):
    """Plots the given 'metric' from 'history'.
    Arguments:
        history: history attribute of History object returned from Model.fit.
        metric: Metric to plot, a string value present as key in 'history'.
        title: A string to be used as title of plot.
        has_valid: Boolean, true if valid data was passed to Model.fit else false.
    Returns:
        None.
    """
    plt.plot(history[metric])
    if has_valid:
        plt.plot(history["val_" + metric])
        plt.legend(["train", "validation"], loc="upper left")
    plt.title(title)
    plt.ylabel(metric)
    plt.xlabel("epoch")
    plt.show()
```

## Visualize pairs and their labels

```python
def visualize(pairs, labels, to_show=6, num_col=3, predictions=None, test=False):
    """Creates a plot of pairs and labels, and prediction if it's test dataset.
    Arguments:
        pairs: Numpy Array, of pairs to visualize, having shape
               (Number of pairs, 2, 28, 28).
        to_show: Int, number of examples to visualize (default is 6)
                `to_show` must be an integral multiple of `num_col`.
                 Otherwise it will be trimmed if it is greater than num_col,
                 and incremented if if it is less then num_col.
        num_col: Int, number of images in one row - (default is 3)
                 For test and train respectively, it should not exceed 3 and 7.
        predictions: Numpy Array of predictions with shape (to_show, 1) -
                     (default is None)
                     Must be passed when test=True.
        test: Boolean telling whether the dataset being visualized is
              train dataset or test dataset - (default False).
    Returns:
        None.
    """

    # Define num_row
    # If to_show % num_col != 0
    #    trim to_show,
    #       to trim to_show limit num_row to the point where
    #       to_show % num_col == 0
    #
    # If to_show//num_col == 0
    #    then it means num_col is greater then to_show
    #    increment to_show
    #       to increment to_show set num_row to 1
    num_row = to_show // num_col if to_show // num_col != 0 else 1

    # `to_show` must be an integral multiple of `num_col`
    #  we found num_row and we have num_col
    #  to increment or decrement to_show
    #  to make it integral multiple of `num_col`
    #  simply set it equal to num_row * num_col
    to_show = num_row * num_col

    # Plot the images
    fig, axes = plt.subplots(num_row, num_col, figsize=(5, 5))
    for i in range(to_show):

        # If the number of rows is 1, the axes array is one-dimensional
        if num_row == 1:
            ax = axes[i % num_col]
        else:
            ax = axes[i // num_col, i % num_col]

        ax.imshow(tf.concat([pairs[i][0], pairs[i][1]], axis=1), cmap="gray")
        ax.set_axis_off()
        if test:
            ax.set_title("True: {} | Pred: {:.5f}".format(labels[i], predictions[i][0]))
        else:
            ax.set_title("Label: {}".format(labels[i]))
    if test:
        plt.tight_layout(rect=(0, 0, 1.9, 1.9), w_pad=0.0)
    else:
        plt.tight_layout(rect=(0, 0, 1.5, 1.5))
    plt.show()
```

## Create pairs of images
We will train the model to differentiate between digits of different classes. For
example, digit `0` needs to be differentiated from the rest of the
digits (`1` through `9`), digit `1` - from `0` and `2` through `9`, and so on.
To carry this out, we will select N random images from class A (for example,
for digit `0`) and pair them with N random images from another class B
(for example, for digit `1`). Then, we can repeat this process for all classes
of digits (until digit `9`). Once we have paired digit `0` with other digits,
we can repeat this process for the remaining classes for the rest of the digits
(from `1` until `9`).

```python
def make_pairs(x, y):
    """Creates a tuple containing image pairs with corresponding label.
    Arguments:
        x: List containing images, each index in this list corresponds to one image.
        y: List containing labels, each label with datatype of `int`.
    Returns:
        Tuple containing two numpy arrays as (pairs_of_samples, labels),
        where pairs_of_samples' shape is (2len(x), 2,n_features_dims) and
        labels are a binary array of shape (2len(x)).
    """

    num_classes = max(y) + 1
    digit_indices = [np.where(y == i)[0] for i in range(num_classes)]

    pairs = []
    labels = []
    pairs_answers = []

    for idx1 in range(len(x)):
        # add a matching example
        x1 = x[idx1]
        label1 = y[idx1]
        idx2 = random.choice(digit_indices[label1])
        x2 = x[idx2]

        pairs += [[x1, x2]]
        labels += [1]

        # add a non-matching example
        label2 = random.randint(0, num_classes - 1)
        while label2 == label1:
            label2 = random.randint(0, num_classes - 1)

        idx2 = random.choice(digit_indices[label2])
        x2 = x[idx2]

        pairs += [[x1, x2]]
        labels += [0]
        pairs_answers += [[label1,label2]]

    return np.array(pairs), np.array(labels).astype("float32"), np.array(pairs_answers) 


# make train pairs
pairs_train, labels_train, pairs_train_answer = make_pairs(x_train, y_train)

# make validation pairs
pairs_val, labels_val, pairs_val_answer = make_pairs(x_val, y_val)

# make test pairs
pairs_test, labels_test, pairs_test_answer = make_pairs(x_test, y_test)
```

# 孿生網路 Siamese Network

```python
class EmbeddingNet(layers.Layer):
    """CNN plays like a Feature Extactor.
    Arguments:
      ouput_dim: int (default None), the output embedding dimension.
    Returns:
      embedding: tensor shape=(batch, ouput_dim)
        if ouput_dim=None, shape=(None, Flattens of the cnn's latent outputs)
    
    Modified from https://github.com/adambielski/siamese-triplet/blob/master/networks.py
    """
    """TF subcalss style.
    * subcalss style 的好處是可以用物件方式方便管理\提取網路層
    * 可以更便利地拆裝不同隱藏層模組，輸入層可透過call自動連接
    * 可以輕鬆將 keras fit 更替為迴圈方式計算梯度
    * override __init__(): 此處定義層 
    * override call(): 此處以 functoinal api 方式設定 forwarding 方式
    """
    
    def __init__(self, ouput_dim:int=None, name="EmbeddingNet", **kwargs):
        super().__init__(name=name, **kwargs)
        """super() functon explanation:
        The super() builtin returns a proxy object ( 
        temporary object of the superclass) that allows us to 
        access methods of the base class.
        
        在python中，繼承父類別的方式只需「將父類別作為子類別初始參數」
        即於定義子類別時，括號內填入父類別名稱。
        如此，子類別已繼承父類別所有功能，
        可視為子類別生成物件(子)時，會呼叫父類別生成物件(父)，
        並由子的物件指向父的物件。
        
        然而當我們需要覆寫(override) __init__ 又想延續原本父類別的內容，
        就會需要用到 super() 將父類別的函式內容、物件內全域變數繼承回來。
        
        super()是class()內建函式，用於呼叫父類別並呼叫其內部函式，
        此為 EmbeddingNet_adambielski 生成物件時，
        EmbeddingNet_adambielski的__init__()自動啟動，
        並接續生成父類別物件，並呼叫其__init__()補完父類別的內部設定，
        使子類別完整繼承其功能
        
        `Python2.x` need to ues `super(child_class_name, self).xxx_functinon()`
        """
        self.output_dim = ouput_dim
        self.Conv_1 = layers.Conv2D(filters = 32, kernel_size=5, activation='PReLU')
        self.MaxPool_1 = layers.MaxPooling2D(pool_size=2, strides=2)
        self.Conv_2 = layers.Conv2D(filters = 64, kernel_size=5, activation='PReLU')
        self.MaxPool_2 = layers.MaxPooling2D(pool_size=2, strides=2)
        self.Flatten = layers.Flatten()
        
        if self.output_dim:
            self.FC_1 = layers.Dense(256, activation='PReLU')
            self.FC_2 = layers.Dense(256, activation='PReLU')
            self.FC_to_outputdim = layers.Dense(self.output_dim, activation=None)
        
    def call(self, inputs):
        """forwarding"""
        x = self.Conv_1(inputs)
        x = self.MaxPool_1(x)
        x = self.Conv_2(x)
        x = self.MaxPool_2(x)
        embedding = self.Flatten(x)
        
        if self.output_dim:
            x = self.FC_1(embedding)
            x = self.FC_2(x)
            embedding = self.FC_to_outputdim(x)
        
        return embedding
    
    def build_graph(self, input_shape:tuple, name=None):
        """use the present layers to construct the model object, 
        then summary() is allowed.
        Arguments: input_shape:tuple
        """
        input_layer = keras.Input(shape=input_shape)
        return Model(inputs=input_layer, outputs=self.call(input_layer), name=name)
    
    def get_embedding(self, inputs):
        """inference"""
        return self.call(inputs)
```

```python
class ClassificationNet(layers.Layer):
    """Shallow Network as mlp classifier.
    Arguments:
      EmbeddingNet: layers.Layer
      num_classes: int 
    Returns:
      class_probilities: tensor shape=(batch, num_classes), 
        the probability distribution of classes.
    
    Modified from https://github.com/adambielski/siamese-triplet/blob/master/networks.py
    """
    
    def __init__(self, EmbeddingNet, num_classes:int, name="ClassificationNet", **kwargs):
        """define layers"""
        super().__init__(name=name, **kwargs)
        self.num_classes = num_classes
        self.EmbeddingNet = EmbeddingNet
        self.nonlinear_activation = layers.PReLU()
        self.FC_softmax_to_classes = layers.Dense(self.num_classes, activation='softmax')
        '''classification layer (mlp+softmax)'''
        
    def call(self, inputs):
        """forwarding"""
        embedding = self.EmbeddingNet(inputs)
        x = self.nonlinear_activation(embedding)
        class_probilities = FC_softmax_to_classes(x)
        return class_probilities
    
    def build_graph(self, input_shape:tuple, name=None):
        """use the present layers to construct the model object, 
        then summary() is allowed.
        Arguments: input_shape:tuple
        """
        input_layer = keras.Input(shape=input_shape)
        return Model(inputs=input_layer, outputs=self.call(input_layer), name=name)
    
    def get_embedding(self, inputs):
        """latent embedding after nonlinear activation"""
        return self.nonlinear_activation(self.EmbeddingNet(inputs))
```

```python
class SiameseNet(keras.Model):
    """
    Arguments:
      EmbeddingNet: layers.Layer
    Returns:
      embedding_1, embedding_2
    
    Modified from https://github.com/adambielski/siamese-triplet/blob/master/networks.py
    """
    
    def __init__(self, EmbeddingNet, name="SiameseNet", **kwargs):
        """define layers"""
        super().__init__(name=name, **kwargs)
        self.EmbeddingNet = EmbeddingNet
    
    def call(self, two_inputs):
        """forwarding"""
        inputs_1, inputs_2 = two_inputs
        embedding_1 = self.EmbeddingNet(inputs_1)
        embedding_2 = self.EmbeddingNet(inputs_2)
        #return embedding_1, embedding_2 # Keras AutoGraph loss function can't recive 2 outputs,  
        output_square_distance = layers.Lambda(self.__square_euclidean_distance, name='distance_layer')([embedding_1, embedding_2])
        return output_square_distance
    
    def build_graph(self, input_shape:tuple, name=None):
        """use the present layers to construct the model object, 
        then summary() is allowed.
        Arguments: input_shape:tuple
        """
        input_layer_1 = keras.Input(shape=input_shape)
        input_layer_2 = keras.Input(shape=input_shape)
        return Model(inputs=[input_layer_1, input_layer_2], outputs=self.call([input_layer_1, input_layer_2]), name=name)
    
    def get_embedding(self, inputs):
        """input one tensor, output its embedding"""
        return self.EmbeddingNet(inputs)
    
    def __square_euclidean_distance(self, vecs):
        '''
        Auther: https://github.com/adambielski
        # square distance
        '''
        output1, output2 = vecs
        square_distances = tf.math.reduce_sum(tf.math.square(output1 - output2), axis=1, keepdims=True)  # squared distances

        return square_distances #tf.math.sqrt(tf.math.maximum(square_distances, tf.keras.backend.epsilon()))
```

## Contractive Loss

```python
def ContrastiveLoss(margin=1.0, eps=1e-9, size_average=True):
    """Contrastive loss
    Takes embeddings of two samples. 
    target label == 1 , if samples are from the same class;
    target label == 0 , otherwise.
    
    Modified from https://github.com/adambielski/siamese-triplet/blob/master/Experiments_MNIST.ipynb
    """
    def loss_fn(y_true, y_pred):
        """
        y_pred: square_distances
        """
        y_true = tf.cast(y_true, tf.float32) # y_true default 1 or 0
        margin_square = tf.math.square(tf.nn.relu(margin - tf.sqrt(y_pred + eps)))
        # margin_square = tf.math.square(tf.math.maximum(margin - (square_distances), 0))
        
        losses = 0.5 * y_true * y_pred + 0.5 * (1 - y_true) * margin_square 
        
        return tf.math.reduce_mean(losses) if size_average else tf.math.reduce_sum(losses)

    return loss_fn
```

```python
def ContrastiveLoss_tfa_Kuihao(margin=1.0):
    def loss_fn(y_true, y_pred):
        '''
        y_pred: square distance
        '''
        y_pred = tf.math.sqrt(tf.math.maximum(y_pred, tf.keras.backend.epsilon()))
        #y_pred = tf.linalg.norm(y_pred, axis=1)
        return tfa.losses.contrastive_loss(y_true,y_pred,margin=margin)
    return loss_fn
```

## 計算特徵中心點

```python
def ClassExtractor_getIndices(data_x, data_y):
    num_classes = max(data_y) + 1
    digit_indices = [np.where(data_y == i)[0] for i in range(num_classes)]
    return np.array(digit_indices)
```

```python
# 計算同類別資料的中心點
def Get_Classes_Mean_Embeddings(embeddings, labels): # Classes=Ways
    # 取得訓練資料分類標籤
    class_indices = ClassExtractor_getIndices(embeddings, labels)
    
    # 計算同類別資料的中心點
    number_of_classes = max(labels) + 1
    class_center_points = [] # np.zeros(number_of_classes) #setting an array element with a sequence.
    for i in range(number_of_classes):
        specific_class_embeddings = embeddings[class_indices[i]]
        class_center_points.append(np.mean(specific_class_embeddings,axis=0))
    #class_center_points = np.array(class_center_points)
    class_center_points = tf.convert_to_tensor(class_center_points)
    
    return class_center_points
```

## 距離預測x10+Softmax

```python
def Similarity(vecs):
    '''negative_euclidean_distance'''
    output1, output2 = vecs
    square_distances = tf.math.reduce_sum(tf.math.square(output1 - output2), axis=1, keepdims=True)  # squared distances
    distances = tf.math.sqrt(tf.math.maximum(square_distances, tf.keras.backend.epsilon()))
    negative_distances = tf.math.multiply(distances,-1) # !!! 將距離取反，將距離轉相似度
    return negative_distances

# tf.keras.losses.CosineSimilarity() #若用此計算，會乘上-1
def Get_Cosine_Similarity(x,y):
    return np.dot(x, y)/(np.linalg.norm(x)*np.linalg.norm(y))

def SiameseNet_inference_Struct(EmbeddingNet, input_shape, num_classes, CenterPoint_inputs):
    EmbeddingNet.trainable = False
    
    input_tensor = Input(shape=input_shape)
    #作圖用
    #CenterPoint_input_tensor = Input(tensor=CenterPoint_inputs)
    
    embedding = EmbeddingNet(input_tensor)
    
    distance_outputs = []
    for i in range(num_classes):
        distance_outputs.append(layers.Lambda(Similarity)([embedding, CenterPoint_inputs[i]]))
        #作圖用
        #distance_outputs.append(layers.Lambda(Similarity)([embedding, CenterPoint_input_tensor[i]]))

    distance_outputs = tf.concat(distance_outputs,1)
    softmax_outputs = tf.nn.softmax(distance_outputs, axis=1)
    
    #作圖用
    #model = Model(inputs=[input_tensor,CenterPoint_input_tensor], outputs=softmax_outputs, name="SiameseNet_inference_Struct")
    model = Model(inputs=input_tensor, outputs=softmax_outputs, name="SiameseNet_inference_Struct")
    
    print(model.summary())
    
    return model
    
SiameseNet_inference = SiameseNet_inference_Struct(Embedding_Net_model, (28,28,1), 10, train_class_center_points)

#作圖用
#tf.keras.utils.plot_model(SiameseNet_inference, to_file='tmp.png', show_shapes=False, show_layer_names=False, dpi=100)
```

```python
class SiameseNet_LoopTrain_Only(keras.Model):
    """
    Arguments:
      EmbeddingNet: layers.Layer
    Returns:
      embedding_1, embedding_2
    
    Modified from https://github.com/adambielski/siamese-triplet/blob/master/networks.py
    """
    
    def __init__(self, EmbeddingNet, name="SiameseNet", **kwargs):
        """define layers"""
        super().__init__(name=name, **kwargs)
        self.EmbeddingNet = EmbeddingNet
    
    def call(self, two_inputs):
        """forwarding"""
        inputs_1, inputs_2 = two_inputs
        embedding_1 = self.EmbeddingNet(inputs_1)
        embedding_2 = self.EmbeddingNet(inputs_2)
        return embedding_1, embedding_2 # Keras AutoGraph loss function can't recive 2 outputs,  
        #output_square_distance = layers.Lambda(self.__square_euclidean_distance, name='distance_layer')([embedding_1, embedding_2])
        #return output_square_distance
    
    def build_graph(self, input_shape:tuple, name=None):
        """use the present layers to construct the model object, 
        then summary() is allowed.
        Arguments: input_shape:tuple
        """
        input_layer_1 = keras.Input(shape=input_shape)
        input_layer_2 = keras.Input(shape=input_shape)
        return Model(inputs=[input_layer_1, input_layer_2], outputs=self.call([input_layer_1, input_layer_2]), name=name)
    
    def get_embedding(self, inputs):
        """input one tensor, output its embedding"""
        return self.EmbeddingNet(inputs)
    
    def __square_euclidean_distance(self, vecs):
        '''
        Auther: https://github.com/adambielski
        # square distance
        '''
        output1, output2 = vecs
        square_distances = tf.math.reduce_sum(tf.math.square(output1 - output2), axis=1, keepdims=True)  # squared distances

        return square_distances #tf.math.sqrt(tf.math.maximum(square_distances, tf.keras.backend.epsilon()))
```

```python
def ContrastiveLoss_tfa_Kuihao_for_LoopTraining(y_true, y_pred_vecs, margin=1.0, eps=1e-9):
    
    vec_x, vec_y = y_pred_vecs
    square_distances = tf.math.reduce_sum(tf.math.square(vec_x - vec_y), axis=1, keepdims=True)
    y_pred = tf.math.sqrt(tf.math.maximum(square_distances, tf.keras.backend.epsilon()))
    losses = 0.5*tfa.losses.contrastive_loss(y_true,y_pred,margin=margin)
    
    '''
    y_true = tf.cast(y_true, tf.float32) # y_true default 1 or 0
    vec_x, vec_y = y_pred_vecs
    square_distances = tf.math.reduce_sum(tf.math.square(vec_x - vec_y), axis=1, keepdims=True)
    margin_square = tf.math.square(tf.nn.relu(margin - tf.sqrt(square_distances + eps)))
    losses = 0.5 * y_true * square_distances + 0.5 * (1 - y_true) * margin_square 
    losses = tf.math.reduce_mean(losses)
    '''
    return losses
    
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((x_train_1[:,:,:,tf.newaxis], x_train_2[:,:,:,tf.newaxis], labels_train))
train_dataset = train_dataset.shuffle(buffer_size=4096).batch(128)
```

```python
optimizer  = tf.keras.optimizers.Adam() #tf.keras.optimizers.Adamax() #tf.keras.optimizers.Adam()
loss_metric = tf.keras.metrics.Mean()
```

```python
E = 10 # E = epochs

# Iterate over epochs.
for epoch in range(E):
    print("Start of epoch %d" % (epoch,))

    # Iterate over the batches of the dataset.
    for step, batch_data in enumerate(train_dataset):
        x_batch_train_1, x_batch_train_2, y_batch_train = batch_data
        with tf.GradientTape() as tape:
            # Get model y_predicts
            two_embeddings = Siamese_Net_LoopTraining_model([x_batch_train_1, x_batch_train_2], training=True)
            # Compute loss (y_predicts vs y_truth)
            loss = ContrastiveLoss_tfa_Kuihao_for_LoopTraining(y_batch_train, two_embeddings)
            loss += sum(Siamese_Net_LoopTraining_model.losses)  
            
        grads = tape.gradient(loss, Siamese_Net_LoopTraining_model.trainable_weights)
        optimizer.apply_gradients(zip(grads, Siamese_Net_LoopTraining_model.trainable_weights))

        loss_metric(loss)

        #if step % 100 == 0:
        #    print("step %d: mean loss = %.4f" % (step, loss_metric.result()))
    print("step %d: mean loss = %.4f" % (step, loss_metric.result()))

print("finish.")
```

```python
def Similarity(vecs):
    '''euclidean_distance'''
    output1, output2 = vecs
    square_distances = tf.math.reduce_sum(tf.math.square(output1 - output2), axis=1, keepdims=True)  # squared distances
    distances = tf.math.sqrt(tf.math.maximum(square_distances, tf.keras.backend.epsilon()))
    #negative_distances = tf.math.multiply(distances,-1) # !!! 將距離取反，將距離轉相似度
    return 1-distances

predictions = Siamese_Net_LoopTraining_model.predict([x_test_1, x_test_2])
similaritys = Similarity(predictions)
# 相似度 = 1 - 預測距離
visualize(pairs_test, labels_test, to_show=10, predictions=similaritys, test=True)
```
